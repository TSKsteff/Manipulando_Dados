```python
import cv2 # importando biblioteca de manipulação de imagens 
import numpy as np # importando biblioteca de funções matematicas
import matplotlib.pyplot as plt #importando biblioteca para manipulação de grafico
from google.colab.patches import cv2_imshow #importando um o arquivo especifo nesse caso o imshow
```

```python
steff = cv2.imread('/content/Date/steff b..jpg',1) # ler a imagem 
cv2_imshow(steff) #exibir a imagem
#cv2.waitKey(0) dentro do paretense é passado o valor dos milisegundos de esperan antes de fechar a imagem 
# cv2.destroyAllWindows() -- fechar todas as janelas abertas 
```

```python
#converção basica
Blue, Green, Red = cv2.split(steff)  # as imagens são formadas por tres tipos de cores que são os que nossos olhos conseguem captar 
#nesse caso o azul o verde e vermelho a função cv2.split ele consegue dividir em pixel a quantidade exata de cada um desses canais separadamente

grayscale = (Blue+Green+Red)/3  # guardando a soma de cada dimenção dividida por um numero positivo  na variavel grayscale
# quando maior for o numero a ser dividido pela soma das dimensões, maior fica a ausencia de cor
grayscale = np.array(grayscale, dtype=np.uint8)  # transformo minha variavel em um tipo de array de 1 byte equivalente a 8bits 

cv2_imshow(grayscale) # exposição do minha imagem depois do procedimentos citados ali encima
```

```python
#conversão ponderada
Blue, Green, Red = cv2.split(steff)#as imagens são formadas por tres tipos de cores que são os que nossos olhos conseguem captar 
#nesse caso o azul o verde e vermelho a função cv2.split ele consegue dividir em pixel a quantidade exata de cada um desses canais separadamente

pondered = 0.299*Blue+0.587*Green+0.114*Red # é um codigo imatavel, no caso para termos uma foto preta e branca podemos sempre 
#estar usando e a formula é a mesma 
#multiplicado pelas cores se basea pela porcentagem e dificuldade que nossos olhos conseguem captar cada cor


pondered = np.array(pondered, dtype=np.uint8) # transformo minha variavel em um tipo de array de 1 byte equivalente a 8bits 


cv2_imshow(pondered)  # exposição do minha imagem depois do procedimentos citados ali encima
```

```python
#Histograma da imagem
histogram = np.zeros(640) # Matriz de zeros com a forma, dtype e ordem dadas.
for i in range(pondered.shape[0]):
                                       # prenchendo a matriz 
    for j in range(pondered.shape[1]):
        histogram[pondered[i,j]] += 1

print(histogram[100])
```

```python
x = np.linspace(0,640,640) # Retorne números espaçados uniformemente durante um intervalo especificado
plt.bar(x,histogram)  # criamos uma grafico de barras
plt.xlabel('intensidade'); # nomeando o indice x do grafico(plano cartesiano)
plt.ylabel('frequência'); # nomeando o indice y do grafico(plano cartesiano)
plt.show() # procura todos os objetos de figura atualmente ativos e abre uma anelas interativas que exibem sua figura 
```

```python
plt.hist(pondered.ravel(),640,[0,640]) # Traçando um histograma
plt.show() # procura todos os objetos de figura atualmente ativos e abre uma o janelas interativas que exibem sua figura 
```

```python
histogram_normalized = histogram/(pondered.shape[0]*pondered.shape[1])

x = np.linspace(0,640,640) # Retorne números espaçados uniformemente durante um intervalo especificado
plt.plot(x,histogram_normalized) # realiza a plotagem e da os valores do eixo x e eixo y 
plt.xlabel('intensidade'); # nomeando o indice x do grafico(plano cartesiano)
plt.ylabel('frequência'); # nomeando o indice y do grafico(plano cartesiano)
plt.show() # procura todos os objetos de figura atualmente ativos e abre uma anelas interativas que exibem sua figura
```
